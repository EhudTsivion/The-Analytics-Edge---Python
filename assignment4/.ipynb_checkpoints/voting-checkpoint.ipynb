### Problem 1.3
Q. Build a logistic regression model for voting using the four treatment group. Which of the following coefficients are significant?

```python
from sklearn import linear_model
logreg = linear_model.LogisticRegression()
```

```python
y, X = patsy.dmatrices('voting ~ civicduty + '
                       'hawthorne + '
                       'self + '
                       'neighbors', 
                       data=gerber, return_type='dataframe')
```

```python
logreg.fit(X, y.values[:, 0])
```

Lets use the f_classif function from scikit_learn to check the p-values:
this f_classif doesn't like the X variable generated by patsy. Pherahps there is a more elegant way of doing this

```python
from sklearn.feature_selection import f_classif
f_classif(gerber[['civicduty', 'hawthorne', 'self', 'neighbors']], y.values[:, 0])
```

The p-values are the second array. All seem to be important. Let's varify this using statsmodels

```python
from statsmodels.discrete.discrete_model import Logit
```

```python
mod1 = Logit(y, X).fit()
mod1.summary2()
```

Also statsmodels show that all variables are important

### Problem 1.4
Using a threshold of 0.3, what is the accuracy of the logistic regression model? 

```python
thresh = 0.3
```

Unlike statsmodels, the predict_proba method gives the probability for the first label (0 - not voting) and the probability for the second label (1 - is voting) as 2 by n matrix. We're interested in the probabilty to vote (in this course at least).

```python
predictions = logreg.predict_proba(X)[:, 1]
```

To calculate the accuracy, lets get the confusion matrix

```python
from sklearn.metrics import confusion_matrix
con_matx = confusion_matrix(gerber['voting'].values, predictions > thresh)
```

Now the accuracy:

```python
con_matx.trace()/con_matx.sum()
```

### Problem 1.5
Q. Using a threshold of 0.5, what is the accuracy of the logistic regression model?

A. OK, lets do it all over again...

```python
thresh = 0.5  # set new thresh
con_matx = confusion_matrix(gerber['voting'].values, predictions > thresh)
# the accuracy
con_matx.trace()/con_matx.sum()
```

```python
1 - gerber['voting'].values.mean()
```

Q2. Compute the AUC of the model.

```python
from sklearn import metrics
```

```python
metrics.roc_auc_score(gerber['voting'].values, predictions)
```

A2. this seems like a small number

For fun, I want to plot the ROC curve

```python
fpr, tpr, thresholds = metrics.roc_curve(gerber['voting'].values, predictions)
```

```python
df = pd.DataFrame(dict(fpr=fpr, tpr=tpr))
```

```python
from ggplot import *
```

```python
ggplot(df, aes(x='fpr', y='tpr')) +\
    geom_line() +\
    geom_abline(linetype='dashed') + \
    ggtitle('ROC Curve') + xlab('False positive') + ylab('True positive')
```

### Problem 2.1 + 2.2
Q. Build a CART tree for voting using all data and the same four treatment variables we used before. Plot the tree. What happens, and if relevant, why?

A. Since the Tree method is different in scikit-learn than R's Cart this question is not really transferable. However, the bottom line is similar to prob. 2.2

```python
# let's first generate and fit the tree object

from sklearn import tree
from IPython.display import Image
import pydotplus
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X, y.values[:, 0])
```

```python
# plot the result

dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=list(X),
                                class_names=list(y),
                                filled=True, rounded=True,
                                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())  
```

A. bottom line, neighbor is the first split, civic duty is the last
